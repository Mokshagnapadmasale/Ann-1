```python
dataset.rename(columns={"person_age":"Age",
                        "person_gender":"Gender",
                        "person_education":"Education",
                        "person_income":"Income",
                        "person_emp_exp":"Experience",
                        "person_home_ownership":"Ownership",
                        "loan_amnt":"Loan_amount",
                        "loan_intent":"Loan_intent",
                        "loan_int_rate":"Intrest",
                        "loan_percent_income":"LoanIncome",
                        "cb_person_cred_hist_length":"HistoryLength",
                        "credit_score":"CreditScore",
                        "previous_loan_defaults_on_file":"PreviousLoan",
},inplace=True)
```

```python
dataset.head(1)
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1:]
```

```python
categoricalcols = X.select_dtypes(include="object").columns.tolist()
categoricalcols.remove("PreviousLoan")
categoricalcols
```

```python
print(dataset.Gender.value_counts())
print("--------------------------------")
print(dataset.Education.value_counts())
print("-----------------------------------------")
print(dataset.Ownership.value_counts())
print("------------------------------------------")
print(dataset.Loan_intent.value_counts())
print("--------------------------------------------")
print(dataset.PreviousLoan.value_counts())
```

```python
minmaxcols = X.select_dtypes(exclude="object").columns.tolist()
minmaxcols.remove("Experience")
minmaxcolss = ['Age','Income','Loan_amount','Intrest','Income','HistoryLength','CreditScore'] 
```

```python
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler,StandardScaler,LabelEncoder
onehot = OneHotEncoder(drop="first")
minmax = MinMaxScaler()
label = LabelEncoder()
stanscaler = StandardScaler()
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.head(1)
```

```python
X_train.head(1)
```

```python
X_train["PreviousLoan"] = label.fit_transform(X_train["PreviousLoan"])
X_test["PreviousLoan"] = label.transform(X_test["PreviousLoan"])
X_train.head(1)
```

```python
preprocess = ColumnTransformer(
    [
        ("OneHotEncoder",onehot,categoricalcols),
        ("MinMaxScaler",minmax,minmaxcolss)
    ],remainder="passthrough"
)
```

```python
X_train_sc = preprocess.fit_transform(X_train)
X_test_sc = preprocess.transform(X_test)
X_train_sc = pd.DataFrame(X_train_sc)
X_test_sc = pd.DataFrame(X_test_sc)  
```

```python
X_train_sc.head(1)
```

```python
print(y_train.value_counts())
```

```python
from imblearn.over_sampling import SMOTE
params = {
    0:24507,
    1:9503
}
smote = SMOTE(sampling_strategy=params,random_state=42)
X_train_smote,y_train_smote = smote.fit_resample(X_train_sc,y_train) 
```

ANN IMPLEMENTATION


```python
import keras
import tensorflow as tf
from tensorflow.keras.models import load_model,Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard 
```

```python
model = Sequential([
    Dense(64,activation="relu",input_shape=(X_train_smote.shape[1],)),
    Dense(32,activation="relu"),
    Dense(1,activation="sigmoid")
]
)
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
loss = tf.keras.losses.BinaryCrossentropy() 
```

```python
model.compile(optimizer=opt,loss=loss,metrics=["accuracy"])
```

```python
import datetime
logs = "logs/fit" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
callback = TensorBoard(log_dir=logs,histogram_freq=1)
stoppage = EarlyStopping(monitor="val_loss",patience=10,restore_best_weights=True)
```

```python
history = model.fit(X_train_smote,y_train_smote,validation_data=(X_test_sc,y_test),epochs=100,callbacks=[callback,stoppage])
```
